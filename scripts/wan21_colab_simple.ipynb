# 🎬 Wan2.1 视频生成 - Colab 快速启动

**设置: Runtime → Change runtime type → L4 GPU → Save**

**然后: Runtime → Run all**

```python
# @title ▶️ 一键启动 Wan2.1
# @markdown ---

# @markdown **模型选择 (推荐1.3B for T4):**
MODEL = "1.3B"  # @param ["1.3B", "14B"]

# ============== 执行 ==============
import os, subprocess, time, threading

WORKSPACE = "/workspace"
os.makedirs(WORKSPACE, exist_ok=True)
os.chdir(WORKSPACE)

# 检查GPU
print("🖥️  GPU状态:")
subprocess.run(["nvidia-smi", "--query-gpu=name,memory.free", "--format=csv,noheader"])

# 克隆ComfyUI
if not os.path.exists(f"{WORKSPACE}/ComfyUI"):
    print("📦 克隆ComfyUI...")
    subprocess.run(["git", "clone", "https://github.com/comfyanonymous/ComfyUI.git"], stdout=subprocess.DEVNULL)
    subprocess.run(["pip", "install", "-r", "requirements.txt"], cwd=f"{WORKSPACE}/ComfyUI", stdout=subprocess.DEVNULL)
    print("✅ ComfyUI就绪")
else:
    print("✅ ComfyUI已存在")
```

```python
# @title 🚀 启动 ComfyUI + Wan2.1

# 启动ComfyUI
def start():
    subprocess.Popen(
        ["python3", "main.py", "--listen", "0.0.0.0", "--port", "8188"],
        cwd=f"{WORKSPACE}/ComfyUI",
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
        preexec_fn=os.setsid
    )

threading.Thread(target=start).start()
print("⏳ ComfyUI启动中...")
time.sleep(20)

# 安装穿透
!pip install pyngrok -q
from pyngrok import ngrok

# 设置你的ngrok token (需要去 https://ngrok.com 免费注册)
NGROK_TOKEN = ""  # @param{type:"string"}

if NGROK_TOKEN:
    ngrok.set_auth_token(NGROK_TOKEN)
    url = ngrok.connect(8188, "http")
    print(f"\n🌐 访问地址: {url}")
    print("⚠️  首次使用需要去 ngrok.com 注册获取token")
else:
    print("\n🔗 使用 localtunnel...")
    subprocess.Popen(["npx", "localtunnel", "--port", "8188"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    time.sleep(3)
    !npx localtunnel --port 8188
```
