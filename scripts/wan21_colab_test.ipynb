```python
# @title ▶️ 一键启动
import os, subprocess, time
from IPython.display import display, HTML, clear_output

WORKSPACE = "/workspace"
os.makedirs(WORKSPACE, exist_ok=True)
os.chdir(WORKSPACE)

# 1. 检查GPU
print("🖥️  GPU状态:")
subprocess.run(["nvidia-smi", "--query-gpu=name,memory.free", "--format=csv,noheader"])

# 2. 克隆ComfyUI
if not os.path.exists(f"{WORKSPACE}/ComfyUI"):
    print("📦 克隆ComfyUI...")
    subprocess.run(["git", "clone", "https://github.com/comfyanonymous/ComfyUI.git"], 
                  stdout=subprocess.DEVNULL)
    subprocess.run(["pip", "install", "-r", "requirements.txt"], 
                  cwd=f"{WORKSPACE}/ComfyUI", stdout=subprocess.DEVNULL)
    print("✅ 安装完成")
else:
    print("✅ ComfyUI已存在")
```

```python
# @title 🚀 启动并获取访问链接
import subprocess, threading, time

# 启动ComfyUI
def start():
    os.chdir(f"{WORKSPACE}/ComfyUI")
    subprocess.Popen(
        ["python3", "main.py", "--listen", "0.0.0.0", "--port", "8188"],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
        preexec_fn=os.setsid
    )

threading.Thread(target=start).start()
print("⏳ ComfyUI启动中(约30秒)...")
time.sleep(25)
print("✅ 服务已启动!")

# 安装穿透
!pip install pyngrok -q
from pyngrok import ngrok

# 方法1: ngrok (推荐,更稳定)
print("\n" + "="*50)
print("🌐 访问方式选择:")
print("="*50)
print("\n选项A - ngrok (推荐):")
print("  1. 去 https://ngrok.com 注册免费账号")
print("  2. 复制你的 token")
print("  3. 在下方单元格填入token")
print("\n选项B - localtunnel (无需注册):")
print("  直接运行下一个单元格")
print("="*50)
```

```python
# @title 🔗 方法A: ngrok (填入token后运行)
from pyngrok import ngrok

NGROK_TOKEN = ""  # @param{type:"string"}

if NGROK_TOKEN:
    ngrok.set_auth_token(NGROK_TOKEN)
    url = ngrok.connect(8188, "http")
    display(HTML(f"""
    <div style='padding:25px; background:linear-gradient(135deg,#1a1a2e,#16213e);border-radius:15px;margin:20px 0;'>
    <h2 style='color:#00ff88;margin:0;'>✅ ComfyUI 已就绪!</h2>
    <p style='margin:15px 0;'>访问地址:</p>
    <a href='{url}' target='_blank' style='font-size:24px;color:#00d4ff;text-decoration:none;'>{url}</a>
    <p style='color:#888;margin-top:15px;'>点击链接打开ComfyUI界面</p>
    </div>
"""))
else:
    display(HTML("""
    <div style='padding:20px;background:#1a1a2e;border-radius:10px;'>
    <p style='color:#ff6b6b;'>⚠️ 请在上方填入ngrok token</p>
    <p style='color:#888;'>免费token注册: https://ngrok.com</p>
    </div>
"""))
```
